```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm
```

```python
df = pd.read_csv('books.csv')
df.head()
```

```python
data =[]
for link in tqdm(df['link']):
    res = requests.get(link)
    soup = BeautifulSoup(res.text , 'html.parser')
    book_type= soup.find('ul', class_= 'breadcrumb').find_all('a',)[2].text
    upc =soup.find('table', class_ = 'table table-striped').find_all('td')[0].text
    price_x=soup.find('table', class_ = 'table table-striped').find_all('td')[2].text
    price_i=soup.find('table', class_ = 'table table-striped').find_all('td')[3].text
    tax=soup.find('table', class_ = 'table table-striped').find_all('td')[4].text
    qn =soup.find('table', class_ = 'table table-striped').find_all('td')[5].text
    reviews=  soup.find('table', class_ = 'table table-striped').find_all('td')[6].text
    data.append([book_type , price_x, price_i, tax , qn , reviews])
```

```python
df =pd.DataFrame(data , columns=['category' , 'price_e_tax', 'price_i_tax', 'tax','quantity' , 'reviews'])
df.head()
```

```python
df.to_csv('data.csv', index=False)
```
